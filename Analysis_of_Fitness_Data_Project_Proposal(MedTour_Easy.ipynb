**TASK 3**

```python
cycling_data = df_activities[df_activities['Type'] == 'Cycling']
```

```python
cycling_data['Average Heart Rate (bpm)'].mean()
```

```python
df_cycle = df_activities[df_activities['Type'] == 'Cycling'].copy()
```

```python
avg_hr_cycle = df_cycle['Average Heart Rate (bpm)'].mean()
df_cycle['Average Heart Rate (bpm)'].fillna(int(avg_hr_cycle), inplace=True)
```

```python
df_cycle
```

```python
df_run = df_activities[df_activities['Type'] == 'Running'].copy()
df_run.isnull().sum()
```

**TASK 4**

```python
df_run_sorted = df_run.sort_index()
runs_subset_2013_2018 = df_run_sorted.loc['2013-01-01':'2018-12-31']
runs_subset_2013_2018
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
ax[0, 0].plot(df_run.index, df_run['Distance (km)'], label='Distance (km)')
ax[0, 0].set_title('Plot 1')
ax[0, 1].plot(df_run.index, df_run['Duration'], label='Duration')
ax[0, 1].set_title('Plot 2')
ax[1, 0].plot(df_run.index, df_run['Average Pace'], label='Average Pace')
ax[1, 0].set_title('Plot 3')
ax[1, 1].plot(df_run.index, df_run['Calories Burned'], label='Calories Burned')
ax[1, 1].set_title('Plot 4')
plt.tight_layout()
plt.show()
```

**TASK 5**

```python
runs_subset_2015_2018 = df_run.loc['2018-01-01':'2015-01-01']
annual_means = runs_subset_2015_2018.resample('A').mean()
weekly_means = runs_subset_2015_2018.resample('W').mean().resample('W').mean()
weekly_counts_average = df_run['Distance (km)'].resample('W').count().resample('W').mean()
print("Annual Means:")
print(annual_means)
print("\nWeekly Means:")
print(weekly_means)
print("\nAverage Number of Trainings per Week:")
print(weekly_counts_average)
```

**TASK 6**

```python
runs_subset_2015_2018 = df_run['2015':'2018']
runs_distance = runs_subset_2015_2018['Distance (km)']
runs_hr = runs_subset_2015_2018['Average Heart Rate (bpm)']
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 8))
ax1.plot(runs_distance.index, runs_distance, label='Distance (km)')
ax1.set_ylabel('Distance (km)')
ax1.legend()
ax2.plot(runs_hr.index, runs_hr, label='Heart Rate (bpm)', color='red')
ax2.set_ylabel('Heart Rate (bpm)')
ax2.legend()
avg_hr = runs_hr.mean()
ax2.axhline(avg_hr, color='blue', linewidth=1, linestyle='-.', label='Average HR')
ax2.legend()
plt.xlabel('Date')
plt.suptitle('Distance (km) and Heart Rate (bpm) - 2015 to 2018')
plt.show()
```

**TASK 7**

```python
df_run_subset_2013_2018 = df_run['2013':'2018']
df_run_dist_annual = df_run_subset_2013_2018['Distance (km)'].resample('A').sum()
plt.figure(figsize=(8.0, 5.0))
plt.axhspan(0, 800, color='red', alpha=0.2)
df_run_dist_annual.plot(kind='bar', color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Total Distance (km)')
plt.title('Annual Total Distance (km) - 2013 to 2018')
plt.show()
```

**TASK 8**

```python
import statsmodels.api as sm
df_run_subset_2013_2018 = df_run['2013':'2018']
df_run_dist_wkly = df_run_subset_2013_2018['Distance (km)'].resample('W').bfill()
decomposition = sm.tsa.seasonal_decompose(df_run_dist_wkly, model='additive')
plt.figure(figsize=(12, 5))
plt.plot(df_run_dist_wkly.index, df_run_dist_wkly, label='Observed Distance', color='blue')
plt.plot(decomposition.trend.index, decomposition.trend, label='Trend', color='red')
plt.xlabel('Date')
plt.ylabel('Distance (km)')
plt.title('Observed Distance of Runs with Decomposed Trend - 2013 to 2018')
plt.legend()
plt.show()
```

**TASK 9**

```python
df_run_hr_all = df_run['2015-03':'2018']['Average Heart Rate (bpm)']
zone_names = ['Zone 1', 'Zone 2', 'Zone 3', 'Zone 4', 'Zone 5']
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df_run_hr_all, bins=[100, 130, 140, 150, 160, 200], color='skyblue')
ax.set_xticklabels(zone_names, rotation=-30, ha='left')
ax.set_xlabel('Heart Rate (bpm)')
ax.set_ylabel('Frequency')
plt.title('Customized Heart Rate Distribution - March 2015 to 2018')
plt.show()
```
